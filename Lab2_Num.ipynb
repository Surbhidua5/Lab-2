# **OneNum**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
airbnb=pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv')
airbnb.head()
```

**Histogram**

```python
data = airbnb[airbnb["price"] < 300]

sns.set_theme(style="darkgrid")
plt.hist(data["price"], bins=range(0, 301, 10), color="#69b3a2", alpha=0.9)
plt.title("Night price distribution of Airbnb appartements", size=12)
plt.show()
```

```python
plt.hist(data["price"], bins=range(0, 300, 3), color="#69b3a2", alpha=0.9)
plt.title("Night price distribution of Airbnb appartements", size=12)
plt.show()
```

**Density**

```python
sns.kdeplot(data['price'], bw=10, color='#69b3a2', fill=True, alpha=0.7, label='Bandwidth: 10')
plt.title('Density Plot of Prices')
```

```python
sns.kdeplot(data['price'], bw=2, color='#69b3a2', fill=True, alpha=0.7, label='Bandwidth: 2')
plt.title('Density Plot of Prices')
```

# **TwoNumUnordered**

```python
twonum=pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv')
twonum.head()
```

**Histogram**

```python

# Plot ground living area distribution
plt.hist(twonum['GrLivArea'], bins=50, color='#69b3a2', alpha=0.9)
plt.title("Ground living area distribution", size=12)
plt.xlabel("Area")
plt.show()
# Plot sale price distribution
plt.hist(twonum['SalePrice']/1000, bins=50, color='#69b3a2', alpha=0.9)
plt.title("Sale price distribution", size=12)
plt.xlabel("Sale price (k$)")
plt.show()
```

**Scatter Plot**

```python
plt.scatter(twonum['GrLivArea'], twonum['SalePrice']/1000, color='#69b3a2', alpha=0.8)
plt.title("Ground living area partially explains sale price of apartments", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.show()
```

```python
p = sns.scatterplot(x="GrLivArea", y="SalePrice", data=twonum, color="#69b3a2", alpha=0.8)
p.set(xlabel='Ground living area', ylabel='Sale price (k$)')

p = sns.lmplot(x="GrLivArea", y="SalePrice", data=twonum, line_kws={'color': 'black', 'alpha': 0.8})
p.fig.suptitle("Linear regression")

p = sns.lmplot(x="GrLivArea", y="SalePrice", data=twonum, lowess=True, line_kws={'color': 'black', 'alpha': 0.8})
p.fig.suptitle("Loess")

plt.show()
```

```python
plt.scatter(x=twonum['GrLivArea'], y=twonum['SalePrice']/1000, s=2, color="#69b3a2", alpha=0.8)
plt.xlabel("Ground living area")
plt.ylabel("Sale price (k$)")
plt.title("Dot size", fontsize=12)
plt.show()
```

```python
sns.kdeplot(twonum['GrLivArea'], twonum['SalePrice']/1000, cmap="Blues", shade=True)
plt.xlabel("Ground living area")
plt.ylabel("Sale price (k$)")
plt.title("Density 2d: contour")
plt.show()
```

```python
sns.scatterplot(x="GrLivArea", y="SalePrice", data=twonum, s=20)
plt.xlabel("Ground living area")
plt.ylabel("Sale price (k$)")
plt.title("Density 2d: area")
sns.kdeplot(twonum["GrLivArea"],twonum["SalePrice"], cmap="Blues", shade=True, shade_lowest=False)
plt.show()
```

```python
plt.hexbin(twonum['GrLivArea'], twonum['SalePrice']/1000)
plt.xlabel("Ground living area")
plt.ylabel("Sale price (k$)")
plt.title("Hexbin")
plt.show()
```

```python
plt.hist2d(twonum['GrLivArea'], twonum['SalePrice']/1000)
plt.xlabel("Ground living area")
plt.ylabel("Sale price (k$)")
plt.title("Hist2D")
plt.show()
```

```python
data = num2.tail(10)
sns.lineplot(data=data, x='date', y='value', color="#69b3a2")
plt.fill_between(data['date'], data['value'], color="#69b3a2", alpha=0.5)
plt.scatter(data['date'], data['value'])
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
sns.set_theme(style="darkgrid")
plt.show()
```

```python
# Create a lineplot of value vs date
plt.plot(data['date'], data['value'], color='#69b3a2')

# Add a shaded area between 8000 and the value
plt.fill_between(data['date'], 8000, data['value'], color='#69b3a2', alpha=0.5)

# Add points on top of the line
plt.scatter(data['date'], data['value'], color='#69b3a2')

# Add title, y-label, and grid
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Create a new column "mycolor" based on the condition "value > 7500"
data["mycolor"] = ["yes" if x > 7500 else "no" for x in data["value"]]

# Tail the last 300 rows of the dataframe
data =num2.tail(300)

# Create a fill_between plot using the matplotlib library
plt.fill_between(data["date"], data["value"], 7500, where=data["value"]>7500, color="#69b3a2", alpha=0.5)
plt.fill_between(data["date"], data["value"], 7500, where=data["value"]<=7500, color="#271569", alpha=0.5)

# Add a title, y-axis label, and remove the legend
plt.title("Evolution of Bitcoin price")
plt.ylabel("Bitcoin price ($)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
num3=pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/4_ThreeNum.csv')
num3.head()
```

```python
data = num3.sort_values(by='pop', ascending=False)
data['pop'] = data['pop']/1000000

plt.figure(figsize=(10,6))
sns.scatterplot(x='gdpPercap', y='lifeExp', size='pop', hue='continent', s=100,data=data, alpha=0.7)
plt.xlabel('GDP Per Capita')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs GDP Per Capita by Population and Continent')

plt.show()
```

```python
data["pop"] = data["pop"]/1000000 # convert population to millions
data = data.sort_values("pop", ascending=False)
data["country"] = data["country"].astype("category")

plt.figure(figsize=(8,6))
sns.scatterplot(x="lifeExp", y="pop", size="lifeExp", hue="continent", data=data,alpha=0.7, palette="viridis")
plt.xlabel("GDP per Capita")
plt.ylabel("Population (Millions)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine()
```

```python
data["pop"] = data["pop"]/1000000 # convert population to millions
data = data.sort_values("pop", ascending=False)
data["country"] = data["country"].astype("category")
plt.figure(figsize=(8,6))
sns.scatterplot(x="gdpPercap", y="pop", size="lifeExp", hue="continent", data=data,alpha=0.7, palette="viridis")
plt.xlabel("GDP per Capita")
plt.ylabel("Population (Millions)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine()
```

```python
tmp = num3.copy()
tmp["annotation"] = ["yes" if (x > 5000 and y < 60) or y < 30 or x > 40000 else "no" for x, y in zip(tmp["gdpPercap"], tmp["lifeExp"])]
tmp["pop"] = tmp["pop"]/1000000
tmp = tmp.sort_values("pop", ascending=False)
tmp["country"] = pd.Categorical(tmp["country"])

plt.figure(figsize=(10,8))
sns.scatterplot(x="gdpPercap", y="lifeExp", size="pop", hue="continent", data=tmp, alpha=0.7)
sns.scatterplot(x="gdpPercap", y="lifeExp", data=tmp[tmp["annotation"]=="yes"], s=60, color="black", label="")
for i, txt in enumerate(tmp[tmp["annotation"]=="yes"]["country"]):
    plt.annotate(txt, (tmp[tmp["annotation"]=="yes"]["gdpPercap"].iloc[i], tmp[tmp["annotation"]=="yes"]["lifeExp"].iloc[i]), size=4)
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.legend(title="Continent")
plt.show()

```

# **SeveralNum**

```python
mtcars = pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/6_SeveralNum.csv')
mtcars.head()
```

```python

data = mtcars[["disp", "drat", "hp", "mpg", "qsec", "wt"]]
for i in data:
  print(i)
 


  # Show the histogram of these variables
  # data = data.melt()
  sns.histplot(data=mtcars[i], x=mtcars[i], color="#69b3a2",bins='auto',element='step')
  
  
  plt.show()
```

```python

data = mtcars[["disp", "drat", "hp", "mpg", "qsec", "wt"]]
mtcars['Unnamed: 0']
```

```python
sns.pairplot(data, vars=["mpg", "disp", "drat", "hp", "qsec", "wt"], kind="scatter", plot_kws={'s': 20, 'alpha': 0.5, 'color': '#69b3a2'}, diag_kws={'color': '#69b3a2', 'alpha': 0.5})
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Perform linkage and create dendrogram
Z = linkage(mtcars, method='ward')
dend = dendrogram(Z)

# Color in function of the cluster
plt.figure(figsize=(1, 7))
plt.tight_layout()
dend['color_list'] = ["#69b3a2", "#404080", "orange"]
plt.show()
```

```python
sns.heatmap(mtcars, cmap='YlGnBu', fmt='f')

# show plot
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
data=mtcars.drop(['Unnamed: 0'],axis=1)
pca = PCA(n_components=2)
pca.fit(data)

pca_data = pca.transform(data)
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_df['name'] = mtcars['Unnamed: 0']

sns.scatterplot(x='PC1', y='PC2', data=pca_df, color="#69b3a2")
sns.set_theme(style="whitegrid")
for i, name in enumerate(mtcars['Unnamed: 0']):
    plt.annotate(name, (pca_df.PC1[i], pca_df.PC2[i]), color="#69b3a2")
plt.show()
```
